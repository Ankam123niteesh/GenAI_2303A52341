```python
from tensorflow.keras.metrics import RootMeanSquaredError
from keras.optimizers import RMSprop
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.models import load_model
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/Housing.csv')
data
```

```python
data['mainroad'].unique()
data.replace('yes',1,inplace =True)
data.replace('no',0,inplace =True)
data['furnishingstatus'].unique()
data.replace('furnished',2,inplace =True)
data.replace('semi-furnished',1,inplace =True)
data.replace('unfurnished',0,inplace =True)
data.head()
```

```python
x=data.drop('price',axis=1)
y=data['price']
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(x)
Y = y
X = pd.DataFrame(X_scaled, columns=x.columns) # Assigining X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
model = Sequential()
```

```python
#hidden layers
model.add(Dense(18, input_dim=X.shape[1], activation='swish'))
model.add(Dense(26, activation='swish'))
model.add(Dense(20, activation='swish'))
model.add(Dense(15, activation='swish'))

model.compile(optimizer=RMSprop(), loss='mean_squared_error', metrics=[RootMeanSquaredError()])
```

```python
# Output layer
model.add(Dense(1))
```

```python
# compiling the model
model.compile(optimizer=RMSprop(), loss='mean_squared_error', metrics=[RootMeanSquaredError()])
```

```python
history = model.fit(X_train, Y_train, epochs=200, batch_size=64, validation_data=(X_test, Y_test))
```

```python
# calculating the RMSE
train_loss, train_rmse = model.evaluate(X_train, Y_train)
test_loss, test_rmse = model.evaluate(X_test, Y_test)
print(f"Training RMSE: {train_rmse}")
print(f"Testing RMSE: {test_rmse}")
```

```python
model.save('housing_price_model.h5')
print("Model saved as 'housing_price_model.h5'")

loaded_model = load_model('housing_price_model.h5')
predictions = loaded_model.predict(X_test)
print(predictions[:10])
```
